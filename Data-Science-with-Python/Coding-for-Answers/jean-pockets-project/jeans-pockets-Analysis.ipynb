# Jean pockets

Are women jean pockets ridiculously small? Let's look at the data!

The file `jean-pocket-measurements.csv` shows pocket measurements for 20 popular brands.
Four pairs of jeans from each brand were measured: men's and women's skinny and straight styles. All jeans were designated a 32-inch waistband. 

The following measurements are included:
- `height_front`: maximum front pocket height (cm)
- `height_back`: maximum back pocket height (cm)
- `width_front`: maximum front pocket width (cm)
- `width_back`: maximum back pocket width (cm)

```python
# Import the pandas library for data manipulation
import pandas as pd

# Load the dataset containing jean pocket measurements
df = pd.read_csv('jean-pocket-measurements.csv')

# Display the dataframe to inspect the data
df
```

### Project Ideas

- What is the average difference in pocket `height_front` between women's and men's jeans?

- Is there a significant difference in pocket `height_front` between skinny and straight styles within the same gender?

- How do back pocket sizes compare between women's and men's jeans?

- Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets?



```python
# Calculate the average front pocket height grouped by gender
avg_height = df.groupby('gender')['height_front'].mean()

# Compute the absolute difference in average front pocket height between men and women
height_diff = abs(avg_height['men'] - avg_height['women']).round(3)

# Print the result with formatting
print(f"Average pocket height difference: {height_diff:.2f} units")
```

```python
# Filter front pocket heights by gender and style
women_skinny = df[(df['gender'] == 'women') & (df['style'] == 'skinny')]['height_front']
women_straight = df[(df['gender'] == 'women') & (df['style'] == 'straight')]['height_front']
men_skinny = df[(df['gender'] == 'men') & (df['style'] == 'skinny')]['height_front']
men_straight = df[(df['gender'] == 'men') & (df['style'] == 'straight')]['height_front']

# Import statistical test function
from scipy.stats import ttest_ind

# Perform independent t-test to compare skinny vs straight styles for women
t_stat_women, p_val_women = ttest_ind(women_skinny, women_straight, equal_var=False)

# Perform independent t-test to compare skinny vs straight styles for men
t_stat_men, p_val_men = ttest_ind(men_skinny, men_straight, equal_var=False)

# Print p-values to assess statistical significance
print(f"Women: p-value = {p_val_women:.4f}")
print(f"Men: p-value = {p_val_men:.4f}")
```

```python
# Calculate average back pocket dimensions grouped by gender
back_pocket = df.groupby('gender')[['height_back', 'width_back']].mean()

# Compute back pocket area for each gender
back_pocket['area_back'] = back_pocket['height_back'] * back_pocket['width_back']

# Display the back pocket size comparison
back_pocket

# Calculate absolute and percentage difference in back pocket area
area_diff = abs(back_pocket.loc['men', 'area_back'] - back_pocket.loc['women', 'area_back'])
percent_diff = (area_diff / back_pocket.loc['men', 'area_back']) * 100

# Print the result with formatting
print(f"\nBack pocket area is {percent_diff:.2f}% smaller in women's jeans compared to men's.")
```

```python
# Define your phone's height in cm (adjust this value if needed)
phone_height = 16.0  # Example: Iqoo Neo9 Pro is ~16 cm tall

# Create a new column to check if the phone fits in the back pocket
df['phone_fits_back'] = df['height_back'] >= phone_height

# Calculate the percentage of jeans that fit the phone by gender
fit_percent = df.groupby('gender')['phone_fits_back'].mean() * 100

# Display the results with formatted output
print("📱 Percentage of jeans that fit the phone in back pocket:")
print(f"Women: {fit_percent.get('women', 0):.2f}%")
print(f"Men: {fit_percent.get('men', 0):.2f}%")
```
