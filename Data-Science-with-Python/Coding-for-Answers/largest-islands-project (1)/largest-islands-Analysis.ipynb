# 🏝️ World's Largest Islands Analysis

This notebook explores global island data using Python and pandas. We'll answer:

- What are the 10 largest islands in the tropics?
- Which islands are the largest in each region?
- How does island rank relate to area?
- Which islands are shared by multiple countries?

```python
# Import necessary libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

## 📥 Load Dataset

```python
# Load the dataset containing island information
df = pd.read_csv('largest-islands.csv')

# Display the first few rows to understand the structure
df.head()
```

## 🌴 Top 10 Largest Islands

```python
# Sort the dataset by area in descending order and select the top 10 islands
largest_islands = df.sort_values(by='area', ascending=False).head(10)

# Display the top 10 largest islands
largest_islands
```

## 🌎 Largest Island by Region

```python
# Group the top islands by region and find the largest island in each
largest_island_by_region = largest_islands.groupby('region')[['island', 'area']].max().reset_index()

# Display the result
largest_island_by_region
```

## 📈 Area vs Rank Line Graph

```python
# Sort the dataset by rank in descending order
largest_islands_by_rank = df.sort_values(by='rank', ascending=False)

# Create a line plot with rank on x-axis and area on y-axis
largest_islands_by_rank.plot(x='rank', y='area', kind='line', marker='s')

# Label the axes and add a title
plt.xlabel('Rank')
plt.ylabel('Area (Million Sq.km)')
plt.title('Island Area vs Rank')
plt.grid(True)
plt.show()
```

## 🏳️‍🌈 Islands Shared by Multiple Countries

```python
# Filter islands where the 'countries' column contains a comma (indicating multiple countries)
islands_with_multiple_countries = df[df['countries'].str.contains(",")]

# Display the result
islands_with_multiple_countries
```
