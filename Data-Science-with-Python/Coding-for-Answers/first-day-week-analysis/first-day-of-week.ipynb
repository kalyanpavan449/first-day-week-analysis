# What is the First Day of the Week?

Whether the Gregorian calendar shows Sunday or Monday as the first day of the week depends on where you live.

Do more countries start the week on Sunday or Monday? What about people? What about by continent?

The file `first-day-of-week.csv` shows the first day of the week for each territory. The file `population.csv` shows the population in the year 2020 for each territory in millions, and the file `four-regions.csv` specifies whether each territory is in asia, europe, africa, or the americas. 


```python
import pandas as pd
df = pd.read_csv('first-day-of-week.csv')
df.head()
```

```python
pop = pd.read_csv('population.csv')
pop.head()
```

### Project Ideas

- How many territories show Friday, Saturday, Sunday, and Monday as the `first_day` of the week?

- How many people start the week on Friday, Saturday, Sunday, and Monday?
	- Hint: This will involve a `merge`.

- Which of the `four_regions` predominantly start the week on Sunday? On Monday? Are there any regions that are more divided between Sunday and Monday?
	- Hint: This will also involve a `merge`.

```python
# Count how many countries fall under each 'first_day'

print("Count of countries grouped by first_day:")
print(df.groupby('first_day').count(), "\n")
```

```python
# How many people start the week on Friday, Saturday, Sunday, and Monday?

pop = pd.read_csv('population.csv')

# Merge on 'alpha3' country code
merged = df.merge(pop, on='alpha3')

# Aggregate population by first day of the week
week_day = merged.groupby('first_day').agg({'population': 'sum'}).round(2)
filtered = merged[merged['first_day'].isin(['fri', 'sat', 'sun', 'mon'])]

# Calculate total population for these days
total_people = filtered['population'].sum().round(2)
print(f"Total_people: {total_people} million\n")
```

```python
# Which of the `four_regions` predominantly start the week on Sunday? On Monday? 
# Are there any regions that are more divided between Sunday and Monday?

region = pd.read_csv('four-regions.csv')

# Merge region info with merged dataset
merged_region = merged.merge(region, on='alpha3')

# Group by region and first day of week, summing population
region_merge = merged_region.groupby(['four_regions', 'first_day']).agg({'population': 'sum'}).round(2)

# Filter only for Sunday and Monday
region_summary = region_merge[region_merge.index.get_level_values('first_day').isin(['sun', 'mon'])]

print("Regional population summary (only Sun & Mon):")
print(region_summary)
```
